### 2. Describe how to get the data.

And the datasets' documentation for each column can be downloaded from https://doi.org/10.21949/1529077.

### 3. Describe who produced the data and how.

The dataset was produced by the Federal Aviation Administration (FAA) and was collected manually. While the FAA collects data through sensors, radar, and digital systems for other types of data collection, for this specific dataset it only included data about the runways themselves which was manually collected.
